## Importing Necessary Modules

### Object detection:
        Object Detection is a common Computer Vision problem which deals with identifying and locating object of certain classes in the image. Interpreting the object localisation can be done in various ways, including creating a bounding box around the object.

```python
import cv2
import matplotlib.pyplot as plt
```

## Datasets & Configuration File

```python
config_file = 'C:/Users/Om Misal/Documents/IOT&COMPUTER_VISION_INTERNSHIP/TASK1/Object_Detection/ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
frozen_model = 'C:/Users/Om Misal/Documents/IOT&COMPUTER_VISION_INTERNSHIP/TASK1/Object_Detection/frozen_inference_graph.pb'
```

```python
model = cv2.dnn_DetectionModel(frozen_model,config_file)
```

## Importing Labels 

```python
classLabels = []
file_name = 'C:/Users/Om Misal/Documents/IOT&COMPUTER_VISION_INTERNSHIP/TASK1/Object_Detection/Labels.txt'
with open(file_name,'rt') as fpt:
    classLabels = fpt.read().rstrip('\n').split('\n')
```

## Printing ClassLabels & its Length

```python
print(classLabels)
print(len(classLabels))
```

```python
model.setInputSize(320,320)
model.setInputScale(1.0/127.5)
model.setInputMean(127.5)
model.setInputSwapRB(True)
```

## Image as Input

```python
img = cv2.imread('C:/Users/Om Misal/Documents/IOT&COMPUTER_VISION_INTERNSHIP/TASK1/Object_Detection/photo-1564564321837-a57b7070ac4f.jpg')

plt.imshow(img)
```

## Converting Image from BGR2RGB

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
ClassIndex, confidence, bbox = model.detect(img,confThreshold=0.5)
```

```python
print(ClassIndex)
```

```python
font_scale =3
font = cv2.FONT_HERSHEY_PLAIN
for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidence.flatten(), bbox):
    cv2.rectangle(img, boxes,(255,0,0),2)
    cv2.putText(img,classLabels[ClassInd-1],(boxes[0]+10,boxes[0]+10), font, fontScale=font_scale,color=(0,0,250), thickness=5)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
